```python
import pandas as pd
import numpy as np

```

```python
#we imported ledger entries from 2018 until today
df=pd.read_excel('proje_github/ledger.xlsx')
df=df[['Date','No','Debit']]
df=df.rename(columns={'Debit':'Sale'})
df.head()
```

```python
#customer list is imported
customer=pd.read_excel('proje_github/customers.xlsx')
customer=customer[['No','Name']]
customer.head()
```

```python
#ledger entries and customer list are combined
tablo=pd.merge(df,customer, how='left', on='No')
tablo.head(10)
```

```python
#The month names are added to the table
date=[]
for i in tablo['Date']:
    date.append(i.strftime("%B"))
tablo['Month']=date
tablo.head()
```

```python
#Only 2020 transactions are taken by the following code
import datetime
date_index=[]
for i in tablo['Date']:
    if i>pd.datetime(2020,1,1):
        date_index.append(1)
    else:
        date_index.append(0)
tablo['2020']=date_index
tablo=tablo[tablo['2020']!=0]
tablo.head()
```

```python
tablo=tablo.pivot_table(columns=['Month'],index=['No','Name'],values=['Sale'], aggfunc='sum').fillna(0)

tablo=tablo['Sale'][['January','February','March','April','May','June','July','August']]
tablo.head()
```

```python
#monthly percentage chane in sales per customer are added to the table
tablo['Change1%']=tablo.pct_change(axis=1)['February']
tablo['Change2%']=tablo.pct_change(axis=1)['March']
tablo['Change3%']=tablo.pct_change(axis=1)['April']
tablo['Change4%']=tablo.pct_change(axis=1)['May']
tablo['Change5%']=tablo.pct_change(axis=1)['June']
tablo['Change6%']=tablo.pct_change(axis=1)['July']
tablo['Change7%']=tablo.pct_change(axis=1)['August']
tablo=tablo[['January','Change1%','February','Change2%','March','Change3%','April','Change4%','May','Change5%','June','Change6%','July','Change7%','August']]
tablo=tablo.fillna(0)
tablo.head()
```

```python
#infinite values are removed from the table
tablo['Change1%']=tablo['Change1%'].replace(np.inf, 1)
tablo['Change2%']=tablo['Change2%'].replace(np.inf, 1)
tablo['Change3%']=tablo['Change3%'].replace(np.inf, 1)
tablo['Change4%']=tablo['Change4%'].replace(np.inf, 1)
tablo['Change5%']=tablo['Change5%'].replace(np.inf, 1)
tablo['Change6%']=tablo['Change6%'].replace(np.inf, 1)
tablo['Change7%']=tablo['Change7%'].replace(np.inf, 1)

tablo['Change1%']=tablo['Change1%'].replace(-np.inf, -1)
tablo['Change2%']=tablo['Change2%'].replace(-np.inf, -1)
tablo['Change3%']=tablo['Change3%'].replace(-np.inf, -1)
tablo['Change4%']=tablo['Change4%'].replace(-np.inf, -1)
tablo['Change5%']=tablo['Change5%'].replace(-np.inf, -1)
tablo['Change6%']=tablo['Change6%'].replace(-np.inf, -1)
tablo['Change7%']=tablo['Change7%'].replace(-np.inf, -1)
tablo.head()
```

```python
tablo['Total']=tablo['January']+tablo['February']+tablo['March']+tablo['April']+tablo['May']+tablo['June']+tablo['July']+tablo['August']

tablo=tablo[tablo.sum(axis=1)>0]
tablo.head()
```

```python
tablo['Change1%']=tablo['Change1%'].map(lambda n: '{:.2%}'.format(n))
tablo['Change2%']=tablo['Change2%'].map(lambda n: '{:.2%}'.format(n))
tablo['Change3%']=tablo['Change3%'].map(lambda n: '{:.2%}'.format(n))
tablo['Change4%']=tablo['Change4%'].map(lambda n: '{:.2%}'.format(n))
tablo['Change5%']=tablo['Change5%'].map(lambda n: '{:.2%}'.format(n))
tablo['Change6%']=tablo['Change6%'].map(lambda n: '{:.2%}'.format(n))
tablo['Change7%']=tablo['Change7%'].map(lambda n: '{:.2%}'.format(n))

tablo.head(10)
```
